# Goal of This File

Retrieve the meaningfull data accross 12 stack overflow csv file that I need to build the ML Model to predict canadian salary. Because each file has its own csv structure (column name, type of data, etc). I need to build a unified view to access all this data (data integration).

The columns I have to find: ["Year","OrgSize","Industry","Experience","Title","Country","Salary"]



I will build this massive unified view incrementaly, by dealing with each csv file by year.
Each of the unified view constructed for each csv file will be appended to the file `stack-overflow-canada.csv`.

To Run this file you will need to download the stack overflow survey csv file from 2011 to 2023. And store them in the folder `Canadian Data/Stack Overflow/`

# Imports

```python
import pandas as pd
```

```python
basePath = "../Canadian Data/Stack Overflow/"
canadianDataFileName = "stack-overflow-canada.csv"
countryInterestedIn = "Canada"
```

```python
ColumnToHave = ["Year","OrgSize","Industry","Experience","Title","Country","Salary"]
```

## 2023

```python
data2023 = pd.read_csv(f"{basePath}2023.csv")
columns2023 = ["OrgSize","Industry","YearsCodePro","DevType","Country","ConvertedCompYearly"]
len(data2023.columns)
```

```python
canadaData2023 = data2023[data2023["Country"]=="Canada"][columns2023]
canadaData2023["Year"] = 2023
canadaData2023 = canadaData2023.rename(columns={"YearsCodePro":"Experience", "DevType":"Title", "ConvertedCompYearly":"Salary"})
canadaData2023 = canadaData2023[ColumnToHave]
canadaData2023
```

```python
canadaData2023.to_csv(f'{basePath}{canadianDataFileName}', index=False)
```
